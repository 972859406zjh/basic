```python
%%time
# 连接数据库
conn_string = "host=" + postgres_host + " port=" + postgres_port + " dbname=" + postgres_datebase + \
              " user=" + postgres_user + " password=" + postgres_password
conn = psycopg2.connect(conn_string)
# 创建游标
curs = conn.cursor()

sql_create = '''CREATE TABLE analysis_test.test(
                id serial,
                prod_name varchar(255),
                price numeric(32,2),
                amount int8,
                money numeric(32,2));'''
# 执行SQL命令
curs.execute(sql_create)

#关闭指针和数据库
conn.commit()
curs.close()
conn.close()
```

#### 插入数据

```python
%%time
# 连接数据库
conn_string = "host=" + postgres_host + " port=" + postgres_port + " dbname=" + postgres_datebase + \
              " user=" + postgres_user + " password=" + postgres_password
conn = psycopg2.connect(conn_string)
# 创建游标
curs = conn.cursor()

sql_create = '''CREATE TABLE analysis_test.test(
                id serial,
                prod_name varchar(255),
                price numeric(32,2),
                amount int8,
                money numeric(32,2));'''
sql_insert = '''insert into analysis_test.test (prod_name,price,amount,money) values('apple',29.9,6000,null)'''
sql_select = 'select * from analysis_test.test'
# 执行SQL命令
curs.execute(sql_create)
curs.execute(sql_insert)
curs.execute(sql_select)
# 获取数据
data = curs.fetchall()
df = pd.DataFrame(data, columns=[x[0] for x in curs.description])
#关闭指针和数据库
conn.commit()
curs.close()
conn.close()
df
```

#### 更新数据

```python
conn = psycopg2.connect(conn_string)
curs = conn.cursor()
sql_update = '''update analysis_test.test set money=price*amount'''

# 执行SQL命令
curs.execute(sql_update)
curs.execute(sql_select)
# 获取数据
data = curs.fetchall()
df = pd.DataFrame(data, columns=[x[0] for x in curs.description])
#关闭指针和数据库
conn.commit()
curs.close()
conn.close()
df
```
